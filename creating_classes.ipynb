```python
class Student:
    def __init__(self, name ,collage,qualification):
        self.name=name
        self.collage=collage
        self.qualification=qualification

    def __repr__(self):
        return  f"{self.name},{self.collage},{self.qualification}"
   
```

```python
class Triangle:
    def __init__(self,b=0,h=0):
        self.base=b
        self.height=h


    def area(self):
        return self.base * self.height /2
    
    def __repr__(self):
        return f'Triangle({self.area()})'
```

```python
data = [Triangle(10,5),Triangle(19,5),Triangle(9,5),Triangle(17,5),Triangle(12,5)]
data
```

```python
import os
class reader:
    """Read a file and return the content"""
    def __init__(self,path):
        if not os.path.exists(path):
            raise Exception('File does not exist')
        self.path = path
        self.content = self.read()

    def read(self):
        with open(self.path, 'r') as f:
            return f.read()
        
    def count_word(self):
        return len(self.content.split())
    
    def count_chars(self):
        return len(self.content)
    
    def __str__(self):
        return self.content

    
```

```python
class user:
    def __init__(self, name, age = 21, gender = 'male'):
        self.name = name
        self.age = age
        self.gender = gender

    def __repr__(self):
        return self.name
```

```python
u1 = user('shahid' ,23 , 'male')
u1
```

```python
class student(user):
    def __init__(self,  name ,collage, klass, rollno , age = 21, gender = 'male'):
        super().__init__(name,age,gender)
        self.collage = collage
        self.klass = klass
        self.rollno = rollno

    def info(self):
        print('student details')
        print('name', self.name)
        print('age', self.age)
        print('collage', self.collage)
        print('class', self.klass)
        print('rollno',self.rollno)
        print('*x*'*10)
              
    
```

```python
s1 = student('sagar', 'DDU', 'MSC','391200', age=25 , gender='male')
s1
```

```python
class Mylist(list):
    def __mul__(self,other):
        output = []
        for i,j in zip(self,other):
            output.append(i*j)
        return output
        
    def sum(self):
        return sum(self)
        
    def mean(self):
        return sum(self)/len(self)
```

```python
x = Mylist([1,2,3,4,5])
y = Mylist([10,20,30,40,50])
x * y
```
