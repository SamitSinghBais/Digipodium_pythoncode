```python
def clean_length_column(length):
    if isinstance(length,str):             #check if price is string
       length = length.replace(',', '')    # removes comma
       length = length.strip()             #removes white space
       if length.isnumeric():
           length = int(length)
           return length
       if len(length) == 0:
            return np.nan
    return length
```

```python
df['length'].apply(clean_length_column)
```

```python
df['cleaned length'] = df['length'].apply(clean_length_column)
```

```python
df
```

```python
def cleaned_width_column(width):
    if isinstance(width,str):
        width = width.split()[0]
        return float(width)
    return width
```

```python
df['cleaned_width'] = df['width'].apply(cleaned_width_column)
df
```

```python
df.info()
```

```python
df.sort_values(by='cleaned_width',ascending=False, inplace=True)
df
```
