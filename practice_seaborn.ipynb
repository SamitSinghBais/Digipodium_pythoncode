```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="darkgrid")
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.head()
```

```python
diamonds.info()
```

```python
sns.relplot(data=diamonds, x="clarity", y="carat")
```

```python
sns.relplot(data=diamonds, x="clarity", y="carat",hue = 'table')
```

```python
sns.relplot(
    data=diamonds,
    x="clarity", y="carat", hue="table", style="table"
)
```

```python
sns.relplot(
    data=diamonds,
    x="clarity", y="carat",
    hue="table", palette="ch:r=-.5,l=.75"
)
```
