```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.histplot(iris,x='sepal_length')
```

```python
sns.histplot(iris,x='sepal_length',bins=20, kde=True)#univarriant( one column)
```

```python
sns.kdeplot(iris,x='sepal_length')
```

```python
fig, ax = plt.subplots(figsize=(5,3))
sns.histplot(iris,x='sepal_length',bins=15,kde=True,ax=ax)
plt.savefig('sepal_length',dpi = 300)
```

```python
sns.histplot(iris,x='sepal_width',bins=20, kde=True)#univarriant( one column)
```

```python
sns.kdeplot(iris,x='sepal_width')
```

```python
fig, ax = plt.subplots(figsize=(5,3))
sns.histplot(iris,x='sepal_width',bins=15,kde=True,ax=ax)
plt.savefig('sepal_width',dpi = 300)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
sns.histplot(iris,x='petal_width',bins=15,kde=True,ax=ax)
plt.savefig('petal_width',dpi = 300)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
sns.histplot(iris,x='petal_length',bins=15,kde=True,ax=ax)
plt.savefig('petal_length',dpi = 300)
```

```python
sns.set_palette('rainbow')

sns.histplotfig, ax = plt.subplots(figsize = (20,5), ncols=4)
sns.histplot(iris,x='sepal_length',bins=15,kde=True,ax=ax[0])
sns.histplot(iris,x='sepal_width',bins=15,kde=True,ax=ax[1])
sns.histplot(iris,x='petal_width',bins=15,kde=True,ax=ax[2])
sns.histplot(iris,x='petal_length',bins=15,kde=True,ax=ax[3])
```

```python
sns.set_palette('rainbow')

sns.histplotfig, ax = plt.subplots(figsize = (20,5), ncols=2,nrows=2)
sns.histplot(iris,x='sepal_length',bins=15,kde=True,ax=ax[0,0],color='red')
sns.histplot(iris,x='sepal_width',bins=15,kde=True,ax=ax[0,1],color='green')
sns.histplot(iris,x='petal_width',bins=15,kde=True,ax=ax[1,0],color='yellow')
sns.histplot(iris,x='petal_length',bins=15,kde=True,ax=ax[1,1],color='orange')
```

```python
sns.color_palette('rainbow')
```

```python
sns.barplot(iris,x='species',y='sepal_length')
plt.show()
```

```python
sns.barplot(iris,errorbar=None)
```

```python
sns.set_palette('husl')
sns.scatterplot(iris,x='sepal_length',y='sepal_width',hue='species')
```
